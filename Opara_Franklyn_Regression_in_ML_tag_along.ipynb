```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

```

```python
X = energydf['T2']
y = energydf['T6']

# Splitting df into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
X_train = X_train.values.reshape(-1, 1)
```

```python
linear_model = LinearRegression()
# Fitting the model to the training set
linear_model.fit(X_train, y_train)

# Predicting with the model
predicted_values = linear_model.predict(X_test.values.reshape(-1, 1))

# Calculating the coefficient of determination value
r2 = r2_score(y_test, predicted_values)
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
energydf.drop(columns = ['date', 'lights'], inplace = True)
normal_energydf = pd.DataFrame(scaler.fit_transform(energydf), columns = energydf.columns)

predictor = energydf.drop(columns = ['Appliances'])
target = normal_energydf ['Appliances']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictor, target, test_size = 0.3, random_state = 42)
l_model = LinearRegression()
l_model.fit(X_train, y_train)


predicted_y = l_model.predict(X_test)
from sklearn.metrics import mean_absolute_error #imports mean_absolute_error from sklearn.metrics
meanabsoluteerror = mean_absolute_error(y_test, predicted_y) #returns MAE

```

```python
from sklearn.linear_model import Lasso

def weight(model, features, col_name):
    weights = pd.Series(model.coef_, features.columns).sort_values()
    weights= weights.reset_index()
    weights.columns = ['Features', col_name]
    weights[col_name] = weights[col_name].round(3)
    return weights




```

```python
linearmodelweights = weight(l_model, X_train, 'linear_model')
linearmodelweights
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 0.4)
ridge.fit(X_train, y_train)

ridge_prediction = ridge.predict(X_test) #prediction using ridge regularization


ridgerootmse = np.sqrt(mean_squared_error(y_test, ridge_prediction))



```

```python
ridgerootmse
```

```python
lasso = Lasso(alpha = 0.001)
lasso.fit(X_train, y_train)

lassoweights = weight(lasso, X_train, 'lasso')
lassoweights
```

```python
nonzero = lassoweights.lasso[lassoweights.lasso != 0]
nonzero
```

```python
lassormse = np.sqrt(mean_squared_error(y_test, lasso.predict(X_test)))
lassormse
```
